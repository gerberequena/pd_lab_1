```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Cargar el dataset

```python
raw_data = pd.read_csv('data/dataset_v1.csv')
```

### Graficamos matriz de correlacion para ver que variables tienen impacto

```python
plt.figure(figsize=(7, 6))

sns.heatmap(
     raw_data.corr(),
     annot=True,        # Show the correlation value numbers
     cmap='crest',
     fmt=".2f",         # Format numbers to two decimal places
     linewidths=0.5,    # Add lines between cells
     linecolor='black'
)
```

```python
plt.figure(figsize=(6,3))
plt.scatter(raw_data['MedHouseVal'].index, raw_data['MedHouseVal'])
plt.show()
```

```python
col = raw_data['MedHouseVal']
Q1 = col.quantile(0.25)
Q3 = col.quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
clean_data = raw_data[(col>= lower) & (col <= upper)]
clean_data.to_csv()
```
